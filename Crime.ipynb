```python
import math
def poisson_process(r,t,k):
    return pow((r*t),k)*exp(-r*t)/math.factorial(k)

# look at poisson process at k = 0,1,2,3,
x = linspace(0,60,200)
f1 = poisson_process(.1,x,0)
f2 = poisson_process(.1,x,2)
f6 = poisson_process(.1,x,5)
plot(x, f1)
plot(x, f2, 'r')
plot(x, f6, 'g')
show()

# look at p(0) vs p(>0) as a function of t
f7 = 1-f1
plot(x,f1)
plot(x,f7,'r')
show()

# fix t and r and look at p(k)
k = range(0,10)
f8 = [poisson_process(.1,30,ki) for ki in k] 
plot(k,f8)
show()

# probability density function for the time between two events is
t = linspace(0,10,100)
f9 = r*exp(-r*t)
plot(t,f9)
show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import random

# time incrment in days
dt = 1
crimes_day = .01
crime_rate = crimes_day*dt
t = 0
time_since_last = 0
times = []
while t < 10000:
    # prob of a crime occuring in time period dt is crime_rate
    r = random.random()
    if r <= crime_rate:
        times.append(time_since_last)
        time_since_last = 0
    t += dt
    time_since_last += dt

# the histogram of the data
n, bins, patches = plt.hist(times, 10, normed=1, facecolor='g')


plt.xlabel('Times')
plt.ylabel('Probability')
plt.title('Histogram of time intervals between crimes')
plt.show()
```

```python
# takes in a file with address_key | day | lat | lon and outputs a map from address_key to a SORTED list of days.
def increment(dictionary,key):
    count = dictionary.get(key,0)
    count += 1
    dictionary[key] = count

def map_key_to_days(filename):
    f = open(filename,"r")
    keyToDays = {}
    header = f.readline()
    lastDay = 0
    for line in f.readlines():
        line = line.split("|")
        key = line[0]
        day = int(line[1])
        lat = float(line[2])
        lon = float(line[3])
        if day > lastDay:
            lastDay = day
        if key not in keyToDays:
            keyToDays[key]=[day]
        else:
            keyToDays[key].append(day)
    f.close()
    for days in keyToDays.values():
        days.sort()
    return keyToDays

def calulate_order(key_to_day_map):
    ordermap = {}
    for dayslist in key_to_day_map.values():
        order = len(dayslist)
        increment(ordermap,order)
    return ordermap
        
def extract_order(key_to_day_map,k):
    result = {}
    for key,days in key_to_day_map.iteritems():
        if len(days) == k:
            result[key]=days
    return result

def count_intervals(key_to_day_map):
    interval_counts = {}
    for key,days in key_to_day_map.iteritems():
        intervals = [days[i] - days[i-1] for i in range(1,len(days))]
        for tau in intervals:
            increment(interval_counts,tau)
    return interval_counts

def last_day(key_to_day_map):
    last = 0
    for days in key_to_day_map.values():
        if len(days) > 0:
            m = max(days)
            if m > last:
                last = m
    return last
```

```python
def shortEtAl(keyToDays,r):
    lastday = last_day(keyToDays)
    keyToDays2 = extract_order(keyToDays,2)
    interval_counts = count_intervals(keyToDays2)
    taus = interval_counts.keys()
    taus.sort()
    numDays = lastday + 1
    totalCount = float(sum(interval_counts.values()))
    counts = [interval_counts.get(t)/totalCount for t in taus]
    # so if we wanted to compare the actual risk t days after a burglary to the expected risk if
    # there was no dependence on previous events we could say it was counts/expected
    expected = [2*(numDays-t)/float(pow(numDays,2)) for t in taus]
    relative = [counts[i]/expected[i] for i in range(0,len(counts))]
    plot(taus[0:r],relative[0:r])
    #
    #plot(taus[0:r],expected[0:r])
    show()
    return relative


```

```python
import pandas as pd
def count_crimes(firstday,lastday, days):
    numdays = lastday - firstday
    totalcrimes = len(days)
    background = totalcrimes/float(numdays)
    intervals = [days[i] - days[i-1] for i in range(1,len(days))]
    intervals.append(lastday - days[-1]) # gap from last crime to when we stopped recording data
    daysback_days = {}
    daysback_crime = {} # this is the distribution of intervals
    for t in intervals:
        increment(daysback_crime,t)
        for daysback in range(0,t+1):
            increment(daysback_days,daysback)
    
    daysback_diff = {}
    for interval, daycount in daysback_days.iteritems():
        crime_count = daysback_crime.get(interval,0)
        crimePerDay = float(crime_count)/daycount
        difference = crimePerDay/background
        daysback_diff[interval] = difference
    return daysback_diff
        
def doOther(keyToDays,r):
    lastday = last_day(keyToDays)
    keyToDays2 = extract_order(keyToDays,2)
    rows = []
    for key, days in keyToDays2.iteritems():
        diffmap = count_crimes(0,lastday,days)
        for interval, diff in diffmap.iteritems():
            row = {"Interval":interval,"Diff":diff}
            rows.append(row)
    
    df = pd.DataFrame(rows)
    df2 = df.groupby('Interval').mean()
    toplot = df2.iloc[0:r,0]
    plot(toplot)
    show()
    return df2
```

```python
import random as rd
rd.seed(37)
grid_dim = 200 # crime will happen on a square grid
num_houses = pow(grid_dim,2)

# lets break a day into 4 blocks of 6 hours each (it is assumed that we can't tell if a burglur came back multiple times wihtin a 6 hour window
periodsPerDay = 4

crime_rate = 1/(periodsPerDay*1000.0) # chance of a crime occuring in a given period according to background.
lastday = 1825 # number of days to simulate for


# for each house in the data, keep a list of the days on which crime occurs
keyToDays = {}
for house in range(num_houses):
    keyToDays[house]=[]

for period in range(lastday*periodsPerDay):
    day = period/periodsPerDay
    for house in range(num_houses):
        prob_crime = crime_rate
        prev_crime = keyToDays.get(house)
        if len(prev_crime) > 0:
            days_since = day - prev_crime[-1]
            prob_crime += 100*crime_rate*exp(-days_since/3) # increase is dependent only on last crime (not full sequence)
        if rd.random() < prob_crime:
            keyToDays[house].append(day)


order =  calulate_order(keyToDays)
print order
plot(order.keys(),order.values())
show()
```
