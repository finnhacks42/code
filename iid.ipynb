Lets say I have a variable $x$ sampled I.I.D from some unknown distribution.

I want to predict the value of ${x}_{n}$ based on the previous samples drawn, ${x}_{1}...{x}_{n-1}$ Note: the ordering of the previous samples is irrelevent as they are drawn independently from the same distribution.

Lets assume I take the sqaured error as my error measure.





```python
from scipy.stats import bernoulli
p  = .8
mu = p

R = bernoulli.rvs(p, size=100000)
arange = linspace(0,1,100)
mses = []
for a in arange:
    h = a*mu
    se = [pow(h-r,2) for r in R]
    mse = mean(se)
    mses.append(mse)
plot(arange,mses)
show()

```
