```python
# plot a 2d kde estimate of a set of points
def kde2d(x,y,cont=True,points=False):
    xmin,xmax = min(x),max(x)
    ymin,ymax = min(y),max(y)
    gx,gy = mgrid[xmin:xmax:100j,ymin:ymax:100j]
    positions = vstack([gx.ravel(),gy.ravel()])
    values = vstack([x,y])
    kernel = stats.gaussian_kde(values)
    f = reshape(kernel(positions).T,gx.shape)
    # plot
    xlim(xmin,xmax)
    ylim(ymin,ymax)
    imshow(np.rot90(f), cmap=plt.cm.gist_earth_r, extent=[xmin, xmax, ymin, ymax])
    if contour:
        cset = contour(gx,gy,f)
        #clabel(cset, inline=1, fontsize=10)
        colorbar()
    if points:
        scatter(x, y, c='r', s=35)

    
    
    
```

```python
# Lets try Gradient Descent, We will let W stand for U and H stand for Vt. So we are trying to find W,H s.t M ~ WH

def error(M,W,H):
    E = M - dot(W,H)
    return sum(pow(E,2)) # this is the Frobius norm squared

def alternate_update(M,W,H,W_update,H_update,iterations):
    errors = []
    for i in xrange(iterations):
        errors.append(error(M,W,H))
        H2 = H_update(M,W,H)
        W2 = W_update(M,W,H)
        H = H2
        W = W2
        
    xlabel("iteration")
    ylabel("error")
    title("Convergence")
    plot(range(iterations),errors)
    show()
    M2 = dot(W,H) # final estimate of M
    
    minW,minH = np.min(W),np.min(H)
    print "Non-negative",minW > 0 and minH > 0, "Smallest W",minW,"Smallest H",minH
    print "Final reconstruction error",errors[-1]
    return (M2,W,H)

def plot_factorization(M,M2,U,W):
    # Two subplots, the axes array is 1-d
    f, axarr = plt.subplots(2, sharex=True)
    axarr[0].plot(M)
    axarr[0].set_title('Original & recononstructed data')
    axarr[1].plot(M2)
    show()

    f, axarr = plt.subplots(2, sharex=True)
    axarr[0].plot(U)
    axarr[0].set_title('Original & recononstructed basis vectors')
    axarr[1].plot(W)
    show()


# additive updates ---------------
def updateH(M,W,H,gamma = 0.001):
    Wt = transpose(W)
    a = dot(Wt,M)
    b = dot(Wt,dot(W,H))
    H2 = H +  gamma*(a - b)
    return H2

def updateW(M,W,H,gamma = 0.001):
    Ht = transpose(H)
    a = dot(M,Ht)
    b = dot(W,dot(H,Ht))
    W2 = W + gamma*(a - b)
    return W2    
# ---------------------------------

# Based on Lee, Daniel D., and H. Sebastian Seung. "Algorithms for non-negative matrix factorization." Advances in neural information processing systems. 2001.
# multiplicative updates ---------- this should ensure non-negative factors since we are always muliplying by a positive term
def mUpdateH(M,W,H):
    Wt = transpose(W)
    a = dot(Wt,M)
    b = dot(Wt,dot(W,H))
    H2 = H*(a/b)
    return H2
    
def mUpdateW(M,W,H):
    Ht = transpose(H)
    a = dot(M,Ht)
    b = dot(W,dot(H,Ht))
    W2 = W*(a/b)
    return W2
# ---------------------------------
    
# initialize with random positive values
W = np.random.random(size=(m,k))
H = np.random.random(size=(k,n))

M2,W2,H2 = alternate_update(M,W,H,updateW,updateH,10000)
plot_factorization(M,M2,U,W2)

M3,W3,H3 = alternate_update(M,W,H,mUpdateW,mUpdateH,100)
plot_factorization(M,M3,U,W3)
```

```python
# An alternative approach, instead of doing one update at a time, on each, 
# we could fix 1 and optimize the other all the way, then reverse. - How do I incorporate contraints?

# fix W, minimize error with respect to H
# lets start with your own simple optimization problem, f(x,y) = x^2 +y^2

# minimizing a scalar function of multiple variables

# function maps a vector to a scalar

# so matrices need to be flatterned.

from scipy.optimize import minimize
# the objective function with respect to the matrix H
def objectiveH(h):
    H4 = h.reshape(k,n)
    return error(M,W4,H4)
    
def objectiveW(w):
    W4 = w.reshape(m,k)
    return error(M,W4,H4)

def jacH(h):
    H4 = h.reshape(k,n)
    a = dot(Wt,M)
    b = dot(Wt,dot(W,H))
    J = -2*(a - b)
    return J.flatten()

def jacW(w):
    W4 = w.reshape(m,k)
    a = dot(M,Ht)
    b = dot(W4,dot(H4,Ht))
    J = -2*(a - b)
    return J.flatten()
    

Wt = transpose(W)
Ht = transpose(H)
H4 = H
W4 = W
w = W.flatten()
h = H.flatten()

errors = []
iterations = 5
for i in range(iterations):
    opt = minimize(objectiveW,w,method='BFGS',jac=jacW,options={'disp':True})
    w = opt.x
    W4 = w.reshape(m,k)
    opt = minimize(objectiveH,h,method='BFGS',jac=jacH,options={'disp':True})
    h = opt.x
    H4 = h.reshape(k,n)
    e = error(M,W4,H4)
    errors.append(e)
M4 = dot(W,H)
plot_factorization(M,M4,U,W)

plot(range(iterations),errors)
show()


    




```
