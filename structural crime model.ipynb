```python
import random
import networkx as nx

rcParams['figure.figsize']  = [x*2 for x in default_figure_size]

class Node:
    def __init__(self,x,y):
        self.x = x
        self.y = y
        
    def dist(self,other):
        return sqrt(pow(self.x - other.x,2)+pow(self.y - other.y,2))
    


def buildLayout(nodes):
    layout = {}
    for i, n in enumerate(nodes):
        layout[i] = [n.x,n.y]
    return layout


def draw_graph(A,graph_pos = None, node_size=1, node_color=None, edge_tickness=1, edges=True):
    node_alpha=0.3
    edge_color='blue'
    edge_alpha=0.3
    if node_color == None:
        node_color='blue'
    # create networkx graph
    G=nx.from_numpy_matrix(A) 

    # these are different layouts for the network you may try
    # shell seems to work best
    if not graph_pos:
        graph_pos=nx.shell_layout(G)
    
    # draw graph
    nx.draw_networkx_nodes(G,graph_pos,node_size=node_size, alpha=node_alpha, node_color=node_color)
    if edges:
        nx.draw_networkx_edges(G,graph_pos,width=edge_tickness, alpha=edge_alpha,edge_color=edge_color)
   
    # show graph
    show()


# generates nodes on a circle of radius R
def generateNodes(npoints, radius):
    nodes = []
    for i in range(npoints):
        r = radius*sqrt(random.random())
        theta = 2*pi*random.random()
        x = r*cos(theta)
        y = r*sin(theta)
        nodes.append(Node(x,y))
    return nodes

def plotNodes(nodes):
    xcoords = []
    ycoords = []
    for node in nodes:
        xcoords.append(node.x)
        ycoords.append(node.y)
    return plot(xcoords,ycoords,'.',markersize=5)

# should return a matrix, which I can then invert, and multiply through B for each timestep iteration.
# c must be <= 1
def addLinks(nodes,alpha,R,c):
    n = len(nodes)
    A = zeros((n,n))
    for i in range(n):
        for j in range(n):
            if i < j: # link node i with node j depending on the distance between them
                p = c*pow(float(R)/nodes[i].dist(nodes[j]),alpha)
                if (random.random() < p):
                    A[i,j] = 1
                    A[j,i] = 1
    return A 
                
random.seed(1)
alpha = 1
num_nodes = 200
radius = 100
c = .1
nodes = generateNodes(num_nodes,radius)


A = addLinks(nodes,alpha,radius,c)
I = np.identity(num_nodes)

#print A
zdg = 1/eigvalsh(A)
smallestPos = -1
smallestAbs = -1
for g in zdg:
    if g > 0 and (smallestPos < 0 or g < smallestPos):
        smallestPos = g
    if smallestAbs < 0 or abs(g) < smallestAbs:
        smallestAbs = g


print "smallest positive",smallestPos
print "smallest absolute",smallestAbs


```

With a probability of connection 

$P({A}_{ij} = 1) =  c*(\frac{R}{{d}_{ij}})^{\alpha}$ where  ${d}_{ij} = distance({node}_{i},{node}_{j})$

The density of the graph, defined as the ratio of the number of edges in the graph to the total number possible ($\frac{n(n-1)}{2}$), depends on both $c$ and $\alpha$

```python
n = 50
maxm = n*(n-1)/2.0
mean_d = []
std_d = []
alpha_list = [0,0.1,0.5]
alpha_list.extend(range(1,20))
for alpha in alpha_list:  
    densities = []
    for i in range(100):
        nodes = generateNodes(n,radius)
        B = addLinks(nodes,alpha,radius,c)
        m = sum(B)/2
        d = m/maxm
        densities.append(d)
    mean_d.append(mean(densities))
    std_d.append(std(densities))
    #print "Alpha",alpha,"density",d

plot(alpha_list,mean_d)
title(r"Graph density vs $\alpha$",fontsize=20)
xlabel(r"$\alpha$",fontsize=20)
ylabel("density",fontsize=20)
show()    
```

The standard deviation of the log-crime rate is proportional to the sum of the sqaured entries of a row of the matrix $M^{-1}$

The way the log crime rate is distributed depends crutially on the value of $\gamma$ and its relationship to the eigenvalues of the adjacency matrix of the network.

With $\gamma << 1/{\kappa}_{1}$ where ${\kappa}_{1}$ is the largest eigenvalue of $A$ the crime rate is very homogeneous.

As $\gamma$ approaches $1/{\kappa}_{1}$, the distribution becomes centralized as we would intuatively expect the model to behave

With $\gamma > ~ 1/{\kappa}_{1}$, the distribution appears to become spatially anti-centralized.

Finally with $\gamma >> 1/{\kappa}_{1}$ the crime rate looks homogeneous but spatially random.

```python
# for each node I want to be able to get the nearest X nodes
def caculateNeighbours(nodes):
    result = {}
    for i in range(len(nodes)):
        result[i] = []
        for j in range(len(nodes)):
            d = nodes[i].dist(nodes[j])
            result[i].append(d)
    for i in result:
        js = range(len(nodes))
        pairs = zip(js,result[i])
        #print pairs
        pairs.sort(key=lambda x:x[1])
        neighbours = [p[0] for p in pairs]
        #print pairs
        #print neighbours
        #print "\n"
        result[i] = neighbours[1:]
    return result
```
