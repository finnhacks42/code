```python
import random
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python

npoints = 5000
xlist = []
ylist = []
for i in range(npoints):
    radius = random.random()
    theta = random.uniform(0,2*math.pi)
    x = radius*math.cos(theta)
    y = radius*math.sin(theta)
    xlist.append(x)
    ylist.append(y)
plot(xlist,ylist,'.',markersize=1)
```

```python
def sample_circle(npoints):
    xlist2 = []
    ylist2 = []
    for i in range(npoints):
        r = sqrt(random.random())
        theta = 2*pi*random.random()
        x = r*cos(theta)
        y = r*sin(theta)
        xlist2.append(x)
        ylist2.append(y)
    return (xlist2,ylist2)

n = 100
x,y = sample_circle(n)

plot(x,y,'.',markersize=1)
show()


# now calculate the inter-event distances
distances = []
for i in range(n):
    for j in range(n):
        d_ij = sqrt(pow(x[i]-x[j],2)+pow(y[i]-y[j],2))
        distances.append(d_ij)
n, bins, patches = hist(distances,bins=50)

y = 1+1/math.pi*(2*(pow(bins,2)-1)*arccos(bins/2.0)-bins*(1+pow(bins,2)/2.0)*sqrt(1-pow(bins,2)/4.0))
print y
#plot(bins,y)
#show()
```

```python
def pCDF(r,t):
   return -r*exp(-r*t)

x = linspace(0,10,100)
y = [pCDF(1,t) for t in x] 
plot(x,y)
```

```python

def invCDF(r,x):
    return (-1/r)*log(x/r)

npoints = 10000
tlist = []
for i in range(npoints):
    t = invCDF(1,random.random())
    tlist.append(t)

n,bins,patches = hist(tlist,100,normed=1)
# add a line onto a histogram ...-exp(x)
y = [exp(-x) for x in bins]
plot(bins,y,'r')
show()
# sucess - we have drawn from the desired distribution ...
```

```python
# now lets draw a series of inter-arrival times from the specified distribution

npoints = 1000
events = [0]
total_time = 0
iat_list = []
for i in range(npoints):
    iat = invCDF(1,random.random())
    iat_list.append(iat)
    total_time += iat
    events.append(total_time)

n, bins, patches = hist(iat_list,100,normed=1) 
y = [exp(-x) for x in bins]
plot(bins,y)
show()
    
pairs = []
for i in range(npoints):
    for j in range(npoints):
        if j > i:
            td = events[j] - events[i]
            pairs.append(td)
hist(pairs,100,normed=1)
show()
# hmmm, this doesn't look like a uniform distribution ...
```

```python
# another way of looking at poisson distribution...

def random_sequence(npoints,maxtime):
    #1) generate points from random-uniform

    points = []
    for i in range(npoints):
        points.append(random.uniform(0,npoints))
    
    #2) sort them
    points.sort()
    return points

npoints = 5000
random.seed(1)

points = random_sequence(npoints,npoints)

#3) caclulate the inter-point distances (inter-arrival times)
iat = []
for i in range(1,len(points)):
    iat.append(points[i] - points[i-1])
n, bins, patches = hist(iat,100,normed=1) 
y = [exp(-x) for x in bins]
plot(bins,y)
show()
# and as expected they nicely fit the poisson distribution. 
```
