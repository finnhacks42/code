```python
import numpy.random as npr
from random import seed
seed(37)
npr.seed(37)

def dummy_code(category,num_categories):
    code = [0]*(num_categories)
    code[category]=1
    return code

num_features = 1
num_tasks = 5

# for each category we have a model y = wx + noise
weights = []
noise_sd = []
n_samples = []
for i in range(num_tasks):
    w = uniform(-1,1,num_features+1)
    #w = beta(1,1,num_features+1)
    weights.append(w)
    sd = uniform(0.1,0.4,1)
    noise_sd.append(sd)
    samples = randint(1,500,1)
    n_samples.append(samples)
    
data_rows = sum(n_samples)
columns = num_features+2 +num_tasks# a column for each feature, num_tasks-1 for dummy coding, 1 for category, 1 for target, 
data = zeros((data_rows,columns))
row = 0
for t in range(num_tasks):
    w = weights[t]
    sd = noise_sd[t]
    samples = n_samples[t]
    for j in range(samples):
        # generate a data row for this task
        x = uniform(-1,1,num_features)
        x2 = [1]
        x2.extend(x)
        y = dot(w,x2)+normal(0,sd,1)
        data[row,0:num_features]= x
        data[row,num_features:num_features+num_tasks] = dummy_code(t,num_tasks)
        data[row,num_features+num_tasks]= t
        data[row,num_features+num_tasks+1]= y        
        row +=1
        
train_rows = npr.choice(range(0,data_rows),size=data_rows/2,replace=False)
test_rows = [row for row in range(0,data_rows) if row not in train_rows]
trainY = data[train_rows,-1]
testY = data[test_rows, -1]
test_cats = data[test_rows,-2]
```

```python
from sklearn import linear_model

trainX1 = data[train_rows,0:num_features]
testX1 = data[test_rows,0:num_features]

lm1 = linear_model.LinearRegression()
lm1.fit(trainX1,trainY)
pred1 = lm1.predict(testX1)

print('Coefficients:', lm1.coef_)
print("RSME: %.2f" % sqrt(mean((pred1 - testY) ** 2)))

scatter(testX1[:,0], testY, marker='+', c=test_cats)
if num_features == 1: # otherwise we want to keep all other features fixed at their mean value
    bounds = zeros((2,1))
    bounds[:,0] = [min(testX1[:,0]),max(testX1[:,0])]
    p = lm1.predict(bounds)
    plot(bounds,p)
    
show()
```

```python
trainX2 = data[train_rows,0:num_features+num_tasks-1]
testX2 = data[test_rows,0:num_features+num_tasks-1]
lm2 = linear_model.LinearRegression()
lm2.fit(trainX2,trainY)
pred2 = lm2.predict(testX2)

print('Coefficients:', lm2.coef_)
print("RSME: %.2f" % sqrt(mean((pred2 - testY) ** 2)))
scatter(testX2[:,0], testY, marker='+', c=test_cats)
```

```python
# need to create new features for each combination of the task features with the other features

data_p = zeros((data_rows,num_features+num_tasks+num_features*num_tasks))

for row in range(data_rows):
    data_p[row,0:num_features+num_tasks]=data[row,0:num_features+num_tasks]
    i = 0
    for t in range(num_tasks):
        for f in range(num_features):
            cross_term = data[row,num_features+t]*data[row,f]
            data_p[row,num_features+num_tasks+i] = cross_term
            i +=1

trainX3 = data_p[train_rows,:]
testX3 = data_p[test_rows,:]
lm3 = linear_model.LinearRegression()
lm3.fit(trainX3,trainY)
pred3 = lm3.predict(testX3)
print('Coefficients:', lm3.coef_)
print("RSME: %.2f" % sqrt(mean((pred3 - testY) ** 2)))


```
